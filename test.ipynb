```python
import numpy as np
import math
```

```python
n = 64
m = n
t = 64
k = 2*math.pi/n
```

```python
from main_test import elbmc, elbmhwm, lbmc, lbmhwm
```

```python
test = []
for solver in [elbmc, elbmhwm, lbmc, lbmhwm]:
    test.append(
        solver.initial(
        np.ones((n,n)),
        np.array([[[
            -math.cos(k*x)*math.sin(k*y),
            math.sin(k*x)*math.cos(k*y)
        ] for y in range(n)] for x in range(n)]),
        dx=1/n, dt=1/m**2, max_deltaH=1e-13, nu=1e-5))
    test[-1].show_flow()
```

```python
for solver in test:
    solver.iter(times = t, showlog = False, nohistory = True)
    solver.show_flow()
```
